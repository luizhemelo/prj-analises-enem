## Geração da visualização

```python
fig = px.choropleth_mapbox(
    data_frame=df, 
    geojson=geo_cidades, 
    locations='id', 
    featureidkey='properties.id',
    color='Num_Pretos',
    animation_frame="Ano",
    color_continuous_scale='reds',
    range_color=(0,5000),
    mapbox_style='carto-positron',
    zoom=3,
    center = {"lat":-15.685413,"lon":-53.679588},
    opacity=0.7,
    height=700,
    labels={
        "id": "Código do município",
        "Num_Pretos": "Número de pretos"
    }
)
fig.update_layout(margin={'r':0,'t':0,'l':0, 'b':0})
fig.update_traces(marker_line_width=0.01)

#fig.show()
fig.write_html("../src/vis/vis03.html")
```
